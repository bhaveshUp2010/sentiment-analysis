```python
import emoji
import pandas as pd
from sklearn.model_selection import train_test_split
import re
```

```python
df = pd.read_csv("/content/apple.csv")
```

```python
df.head()
```

```python
def define_lables(df):
  df["label"] = "positive"
  df.loc[df["rating"] == 3, "label"] = "neutral"
  df.loc[df["rating"] < 3, "label"] = "negative"
  return df
```

```python
df = define_lables(df)
```

```python
def clean_text(text):
  text = text.lower()
  text = re.sub(r"\s+", " ", text)
  text = re.sub(r"\s+", " ", text).strip()
  text = re.sub(r"(.)\1{2,}", r"\1\1", text)
  text = emoji.demojize(text)
  return text
```

```python
df["cleaned_text"] =  df['review_text'].apply(clean_text)
```

```python
df = df[df['review_text'].str.len() > 5]
```

```python
train_df, temp_df = train_test_split(
    df,
    test_size = 0.2,
    random_state= 42,
    stratify= df["label"]
)
val_df, test_df = train_test_split(
    temp_df,
    test_size = 0.5,
    random_state = 42,
    stratify = temp_df["label"]
)
```
